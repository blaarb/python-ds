```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset('penguins')
penguins.info()
print(penguins["species"].value_counts())
```

```python
gentoo_flippers = penguins.loc[penguins["species"] == 'Gentoo', "flipper_length_mm"]
# adelie_flippers_sample = list(penguins.loc[penguins["species"] == 'Adelie', "flipper_length_mm"].sample(49))
gentoo_flippers_sample01 = list(gentoo_flippers[:49])
gentoo_flippers_sample02 = list(gentoo_flippers[50:99])
```

```python
compare_df = pd.DataFrame({"gentoo_flippers_sample01": gentoo_flippers_sample01, "gentoo_flippers_sample02": gentoo_flippers_sample02})
compare_df
```

```python
from scipy.stats import ttest_ind
t_stat, p_val = ttest_ind(gentoo_flippers_sample01, gentoo_flippers_sample02)
print(f"t_stat: {t_stat}, p_val: {p_val}")
import numpy as np
print(f"mean для выборки 1: {np.mean(gentoo_flippers_sample01)}\nmean для выборки 2: {np.mean(gentoo_flippers_sample02)}")
```
